```python
### Server Cost Calculator

def get_cost_of_server_per_hour():
    cost_of_server_per_hour = float(input("What is the cost of your server per hour? "))
    return cost_of_server_per_hour

def get_required_hours():
    hours = int(input("How many hours do you need the server for? "))
    return hours

def get_required_days():
    days = int(input("How many days do you need the server for? "))
    return days

def server_cost_calc_hours():  
    cost_per_server_per_hour = get_cost_of_server_per_hour()
    hours = get_required_hours()
    total_cost = hours * cost_per_server_per_hour
    print("The total cost for {} hours is ${}".format(hours, total_cost))

def server_cost_calc_days():
    cost_per_server_per_day = get_cost_of_server_per_hour() * 24
    days = get_required_days()
    total_cost = days * cost_per_server_per_day
    print("The total cost for {} days is ${}".format(days, total_cost))

def server_cost_calculator():
    print("Welcome to the server cost calculator!")
    hours_or_days = input("Would you like to calculate the cost for hours or days? (Enter 'hours' or 'days'): ")
    if hours_or_days.lower() == 'hours':
        server_cost_calc_hours()
    elif hours_or_days.lower() == 'days':
        server_cost_calc_days()
    else:
        print("Invalid input. Please enter 'hours' or 'days'.")

server_cost_calculator()
```

```python
### Saving Lap Times and Deltas
### Using a loop to find how many sectors are in a lap
### Saving the deltas as a new sector begins
### Adding a new Lap once the last lap is complete
### Using an array of key value pairs to save the lap times and deltas for each sector

lap_times = []
deltas = []
lap_number = 1
sector_number = 1
current_lap_time = 0


def get_sector_time(sector_number, lap_number):
    while True:
        text = input(f"Sector {sector_number} time for lap {lap_number}: ".strip())
        if text == "":
            return 0.0
        try:
            return float(text)
        except ValueError:
            print("Please enter a valid number (e.g 12.34 or press Enter to finish the lap)")
        break
        
def add_sector_time_to_lap_time(sector_time):
    global current_lap_time
    current_lap_time += sector_time
    return current_lap_time

def save_lap_times_and_deltas(lap_times, deltas, lap_number):
    lap_data = {
        "lap_number": lap_number,
        "lap_time": sum(deltas),
        "deltas": deltas.copy()
    }
    lap_times.append(lap_data)

sector_time = get_sector_time(sector_number, lap_number)
add_sector_time_to_lap_time(sector_time)

def covert_seconds_to_minutes_and_seconds(seconds):
    minutes = int(seconds // 60)
    remaining_seconds = seconds % 60
    return f"{minutes}m {remaining_seconds:.3f}s"

def lap_timer():
    global sector_number, lap_number, current_lap_time
    while True:
        sector_time = get_sector_time(sector_number, lap_number)
        if sector_time == 0:
            save_lap_times_and_deltas(lap_times, deltas, lap_number)
            print("Lap {} completed with a time of {}".format(lap_number, covert_seconds_to_minutes_and_seconds(sum(deltas))))
            lap_number += 1
            sector_number = 1
            deltas.clear()
            current_lap_time = 0
        else:
            deltas.append(sector_time)
            current_lap_time += sector_time
            print("Sector {} time: {} ".format(sector_number, covert_seconds_to_minutes_and_seconds(sector_time)))
            sector_number += 1
lap_timer()



```

```python
### Text Analysis Tool

def input_paragraph():
    paragraph = input("Enter a paragraph of text: ")
    return paragraph

def analyze_text(paragraph):
    word_count = len(paragraph.split())
    char_count = len(paragraph)
    sentence_count = paragraph.count('.') + paragraph.count('!') + paragraph.count('?')
    vowel_count = sum(paragraph.lower().count(vowel) for vowel in "aeiou")
    consonant_count = sum(paragraph.lower().count(char) for char in "bcdfghjklmnpqrstvwxyz")
    most_common_word = max(set(paragraph.split()), key=paragraph.split().count)
    return word_count, char_count, sentence_count, vowel_count, consonant_count, most_common_word

def print_analysis(word_count, char_count, sentence_count, vowel_count, consonant_count, most_common_word):
    print(f"Word Count: {word_count}")
    print(f"Character Count: {char_count}")
    print(f"Sentence Count: {sentence_count}")
    print(f"Vowel Count: {vowel_count}")
    print(f"Consonant Count: {consonant_count}")
    print(f"Most Common Word: {most_common_word}")

def text_analysis_tool():
    paragraph = input_paragraph()
    analysis = analyze_text(paragraph)
    print_analysis(*analysis)

text_analysis_tool()
```

```python
def input_paragraph():
    paragraph = input("Enter a paragraph: ")
    return paragraph
 
def count_characters(paragraph):
    character_count = len(paragraph)
    return character_count
 
def count_words(paragraph):
    words = paragraph.split()
    word_count = len(words)
    return word_count
 
def count_sentences(paragraph):
    sentences = paragraph.split('.')
    sentence_count = len(sentences)
    return sentence_count
 
def count_vowels(paragraph):
    vowels = 'aeiouAEIOU'
    vowel_count = sum(1 for char in paragraph if char in vowels)
    return vowel_count
 
def count_consonants(paragraph):
    vowels = 'aeiouAEIOU'
    consonant_count = sum(1 for char in paragraph if char.isalpha() and char not in vowels)
    return consonant_count
 
def count_most_frequent_word(paragraph):
    words = paragraph.split()
    word_frequency = {}
    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1
    most_frequent_word = max(word_frequency, key=word_frequency.get)
    return most_frequent_word
 
def main():
    paragraph = input_paragraph()
    print(f"Total characters: {count_characters(paragraph)}")
    print(f"Total words: {count_words(paragraph)}")
    print(f"Total sentences: {count_sentences(paragraph)}")
    print(f"Total vowels: {count_vowels(paragraph)}")
    print(f"Total consonants: {count_consonants(paragraph)}")
    print(f"Most frequent word: {count_most_frequent_word(paragraph)}")
 
if __name__ == "__main__":
    main()
```

```python
import random 

def user_guessing_game():
    guesses = 7
    random_number = random.randint(1, 50)
    while guesses > 7:       
        user_guess = int(input("Guess a number between 1 and 50: "))
        if user_guess < random_number:
            guesses -= 1
            print(f"Guesses left: {guesses}")
            if guesses == 0:
                print("You've run out of guesses. Game over!")
                break
            print("Too low! Try again.")
        elif user_guess > random_number:
            guesses -= 1
            print(f"Guesses left: {guesses}")
            if guesses == 0:
                print("You've run out of guesses. Game over!")
                break
        else:
            print("Congratulations! You guessed the number!")
            break

user_guessing_game()
```

```python
num1 = int(input("Enter a number : "))
num2 = int(input("Enter a second number: "))
operator = input("Would you like to +,-,* or / your numbers?: ")
if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    print(num1 / num2)
elif operator == "/" and num2 == 0:
    print("You cannot divide by zero!")
else:    
    print("Invalid operator")
```

```python
european_countries = ["UK", "France", "Italy", "Germany"]
americas_countries = ["Mexico", "USA", "Canada", "Peru"]
african_countries = ["Kenya", "Ethiopia", "Nigeria", "Ghana"]
asian_countries = ["China", "Japan", "India", "Thailand"]
print(european_countries)
print(americas_countries)
print(african_countries)
print(asian_countries)
european_countries.pop(3)
americas_countries.pop(1)
african_countries.pop(2)
asian_countries.pop(0)
print(european_countries)
print(americas_countries)
print(african_countries)
print(asian_countries)
americas_countries.clear()
print(americas_countries)
del african_countries
americas_countries.append("Brazil")
americas_countries.append("Chile")
americas_countries.extend(["Argentina", "Colombia"])
print(americas_countries)
asian_countries.extend(["Vietnam", "Indonesia"])
print(asian_countries)
```

```python
firsttuple = ("football", "rugby", "cricket", "darts", "football")
print(firsttuple)
print(type(firsttuple))
tuplechanges=list(firsttuple)
print(type(tuplechanges))
# Make changes to the list version of the tuple
tuplechanges.append("tennis")
tuplechanges.remove("darts")
tuplechanges.insert(1, "baseball")
print(tuplechanges)
# Convert back to a tuple
firsttuple = tuple(tuplechanges)
print(firsttuple)
print(type(firsttuple))
```

```python
### Voting Eligibility Checker

user_age = int(input("How old are you? "))
    
def voting_eligibility(user_age):
    if user_age < 18:
        years_left = 18 - user_age
        return("At {} you can vote in {} years!".format(user_age, years_left))
    else:
        return("At {} you are already eligible to vote!".format(user_age))
    
result = voting_eligibility(user_age)
print(result)
```

```python
### Server Cost Calculator

def get_cost_of_server_per_hour():
    cost_of_server_per_hour = float(input("What is the cost of your server per hour? "))
    return cost_of_server_per_hour

def get_required_hours():
    hours = int(input("How many hours do you need the server for? "))
    return hours

def get_required_days():
    days = int(input("How many days do you need the server for? "))
    return days

def server_cost_calc_hours():  
    cost_per_server_per_hour = get_cost_of_server_per_hour()
    hours = get_required_hours()
    total_cost = hours * cost_per_server_per_hour
    print("The total cost for {} hours is ${}".format(hours, total_cost))

def server_cost_calc_days():
    cost_per_server_per_day = get_cost_of_server_per_hour() * 24
    days = get_required_days()
    total_cost = days * cost_per_server_per_day
    print("The total cost for {} days is ${}".format(days, total_cost))

def server_cost_calculator():
    print("Welcome to the server cost calculator!")
    hours_or_days = input("Would you like to calculate the cost for hours or days? (Enter 'hours' or 'days'): ")
    if hours_or_days.lower() == 'hours':
        server_cost_calc_hours()
    elif hours_or_days.lower() == 'days':
        server_cost_calc_days()
    else:
        print("Invalid input. Please enter 'hours' or 'days'.")

server_cost_calculator()
```

```python
### Travel Method

def get_travel_distance():
    distance = int(input("How many miles do you need to travel? "))
    return distance

def calculate_travel_method(distance):
    if distance < 3:
        return("You can walk!")
    elif distance >= 3 and distance < 300:
        return("You can drive!")
    elif distance >= 300 and distance < 40000:
        return("You can fly!")
    else:
        return("You need a rocketship!")
    
def travel_method():
    distance = get_travel_distance()
    method = calculate_travel_method(distance)
    print(method)

travel_method()
```
